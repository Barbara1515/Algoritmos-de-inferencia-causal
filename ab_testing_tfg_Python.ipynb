## 6. Definición del problema causal

El objetivo de este análisis es estimar el efecto causal del envío de un correo electrónico promocional dirigido a productos femeninos sobre el comportamiento posterior de los clientes.

Para ello se comparará el grupo **Womens E-Mail**, que constituye el grupo tratado, con el grupo **No E-Mail**, que actuará como grupo de control.

La variable resultado considerada será inicialmente **conversion**, que indica si el cliente realizó una compra tras la campaña.

Con esta comparación se pretende estimar el **Average Treatment Effect (ATE)** del tratamiento sobre la probabilidad de comprar tras la campaña

### 6.1. Definición de la variable de tratamiento

Sea $T_i$ la variable indicadora de tratamiento correspondiente al individuo $i$, definida como

$$
T_i=
\begin{cases}
1, & \text{si el individuo pertenece al grupo Womens E-Mail},\\
0, & \text{si el individuo pertenece al grupo No E-Mail}.
\end{cases}
$$


```python
# Seleccionamos únicamente los grupos de interés.
df_ab = df[df["segment"].isin(["Womens E-Mail", "No E-Mail"])].copy()
```

```python
# Creamos la variable indicadora del tratamiento
df_ab["T"] = (df_ab["segment"] == "Womens E-Mail").astype(int)
```

```python
# Comprobamos la distribución del tratamiento
df_ab["T"].value_counts()
```

```python
# Visualizamos las primeras observaciones
df_ab[["segment", "T"]].head()
```

### 6.2. Definición de la variable resultado

Sea $Y_i$ la variable resultado correspondiente al individuo $i$.

En este estudio se considera como variable resultado la variable **conversion**, que indica si el cliente realizó una compra o no tras la campaña.

Formalmente,

$$
Y_i=
\begin{cases}
1,& \text{el cliente realizó una compra},\\
0,& \text{en caso contrario}.
\end{cases}
$$


```python
df_ab["conversion"].value_counts()
```

```python
df_ab["conversion"].value_counts(normalize=True)
```


Su distribución permite obtener una primera visión del porcentaje de clientes que realizaron una compra independientemente del grupo experimental al que pertenecen:

```python
plt.figure(figsize=(6,4))

sns.countplot(data=df_ab, x="conversion")

plt.title("Distribución de la variable resultado")
plt.xlabel("Realizan compra")
plt.ylabel("Número de clientes")

plt.show()
```

El gráfico muestra el número de clientes que realizaron o no una compra tras la campaña.

No obstante, esta representación no permite extraer conclusiones causales, ya que no distingue entre los individuos tratados y los pertenecientes al grupo de control.

Para ello será necesario analizar la variable resultado en función del tratamiento recibido. En particular, se calculará la proporción de clientes que realizaron una compra en el grupo tratado y en el grupo de control.

```python
df_ab.groupby("T")["conversion"].mean()
```

Las medias obtenidas representan la proporción de clientes que realizaron una compra tras la campaña en cada uno de los grupos.

Una diferencia apreciable entre ambas proporciones constituye un primer indicio de la posible existencia de un efecto del tratamiento. Sin embargo, para determinar si dicha diferencia es estadísticamente significativa será necesario realizar un contraste de hipótesis.

## 7. Estimación del efecto medio del tratamiento (ATE)

Se procede a estimar el **Average Treatment Effect (ATE)**.

Recordemos que en un experimento aleatorizado, el ATE puede estimarse mediante la diferencia entre la media de la variable resultado en el grupo tratado y la media correspondiente al grupo de control.

```python
# Separamos las observaciones del grupo tratado y del grupo de control
treated = df_ab[df_ab["T"] == 1]
control = df_ab[df_ab["T"] == 0]
```

```python
# Calculamos la probabilidad media de visita en cada grupo
mean_treated = treated["conversion"].mean()
mean_control = control["conversion"].mean()

print(f"Probabilidad media de visita (tratamiento): {mean_treated:.4f}")
print(f"Probabilidad media de visita (control): {mean_control:.4f}")
```

```python
# Estimamos el efecto medio del tratamiento (ATE)
ate = mean_treated - mean_control

print(f"ATE estimado: {ate:.4f}")
```

El ATE estimado de 0,0031 indica que el envío del correo electrónico incrementa la probabilidad de compra en 0,31 puntos porcentuales por cliente. En la muestra analizada, donde aproximadamente 22.000 clientes recibieron el correo, este efecto equivale a unas 68 compras adicionales atribuibles a la campaña.

```python
# Calculamos la tasa media de visitas en cada grupo
grupo = df_ab.groupby("T")["conversion"].mean().reset_index()

# Renombramos los grupos para facilitar la interpretación del gráfico
grupo["Grupo"] = ["Control", "Tratamiento"]
```

```python
# Representamos gráficamente la tasa media de visitas
plt.figure(figsize=(6,4))

plt.bar(grupo["Grupo"], grupo["conversion"])

plt.title("Probabilidad media de visita por grupo")
plt.xlabel("Grupo experimental")
plt.ylabel("Proporción de visitas")

plt.show()
```

## 9. Contraste de hipótesis

La diferencia observada entre las medias de ambos grupos proporciona una estimación del efecto medio del tratamiento. Sin embargo, esta diferencia puede deberse al efecto del tratamiento o simplemente a la variabilidad muestral.

Para evaluar si la diferencia observada es estadísticamente significativa, se realiza un contraste de hipótesis para comparar las medias de la variable resultado entre el grupo tratado y el grupo de control.

Dado que la variable resultado es binaria y ambos grupos son independientes, se emplea un **test t de Student para dos muestras independientes (versión de Welch)**, que no requiere asumir igualdad de varianzas entre los grupos.

Se plantean las siguientes hipótesis:

- **Hipótesis nula ($H_0$)**: el tratamiento no produce efecto sobre la variable resultado, es decir,

$$
ATE = 0.
$$

- **Hipótesis alternativa ($H_1$)**: el tratamiento produce un efecto sobre la variable resultado,

$$
ATE \neq 0.
$$

Si el p-valor obtenido es inferior al nivel de significación establecido ($\alpha = 0.05$), se rechazará la hipótesis nula y se concluirá que existe evidencia estadística de un efecto del tratamiento.

```python
# Importamos la función necesaria para realizar el test t
from scipy.stats import ttest_ind

# Aplicamos el test t de Welch para comparar las medias de ambos grupos
t_stat, p_value = ttest_ind(
    treated["conversion"],
    control["conversion"],
    equal_var=False
    
)

# Mostramos los resultados obtenidos
print(f"Estadístico t: {t_stat:.4f}")
print(f"p-valor: {p_value:.10f}")
```

Como el p-valor obtenido es inferior al nivel de significación establecido ($\alpha = 0.05$), se rechaza la hipótesis nula y se concluye que existe evidencia estadística de un efecto del tratamiento.

## 10. Intervalo de confianza del efecto medio del tratamiento

Además de estimar el efecto medio del tratamiento, resulta de interés cuantificar la incertidumbre asociada a dicha estimación.

Para ello se calcula un intervalo de confianza del 95 % para el ATE. Este intervalo proporciona un rango de valores plausibles para el efecto causal del tratamiento sobre la probabilidad de realizar una compra tras la campaña.

```python
# Calculamos el tamaño muestral de cada grupo
n_t = len(treated)
n_c = len(control)

# Calculamos la varianza de la variable resultado en cada grupo
var_t = treated["conversion"].var()
var_c = control["conversion"].var()

# Calculamos el error estándar del ATE
se = np.sqrt(var_t / n_t + var_c / n_c)

# Calculamos el intervalo de confianza del 95 %
ic_inf = ate - 1.96 * se
ic_sup = ate + 1.96 * se

print(f"Intervalo de confianza (95 %): [{ic_inf:.5f}, {ic_sup:.5f}]")
```

El intervalo de confianza obtenido permite evaluar la precisión de la estimación del efecto medio del tratamiento. Como el intervalo no contiene el valor cero, existe evidencia estadística de que el tratamiento produce un efecto distinto de cero sobre la variable resultado.
